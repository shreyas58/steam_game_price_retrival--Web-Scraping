### Importing necessary libraries

```python
import bs4
import pandas as pd
from datetime import datetime
import re

from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
```

```python
# the URLto scrape
my_url = "https://store.steampowered.com/search/?sort_by=_ASC&filter=topsellers"
```

```python
uClient = uReq(my_url)
page_html = uClient.read()
uClient.close()

page_soup = soup(page_html, "html.parser")

page_soup
```

```python
# storing top 50 game divs
games = page_soup.findAll("div", {"class":"responsive_search_name_combined"})
games[0]
```

```python
# creating a dataframe to store top 50 games
game_prices = pd.DataFrame(columns=['Game_Name'])
game_prices
```

```python
# iterating through each game div to retrieve the game name
for game in games:
    game = str(game)
    game_name = re.search('>.*</span', game).group()
    game_name = game_name[1:-6]
    game_prices = game_prices.append({'Game_Name':game_name}, ignore_index=True)
    
game_prices
```

```python
# storing to a CSV file
game_prices.to_csv('gamePrices.csv', index = False)
```
